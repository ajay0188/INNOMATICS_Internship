```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('aspiring_minds_employability_outcomes_2015.xlsx')
df.head()
```

```python
df.tail()
```

```python
#1.find the correlation coff b/w the complete study % i.e avg 0f all the academics and salary?
df.loc[:,['Salary','ComputerScience','MechanicalEngg','ElectricalEngg','TelecomEngg','CivilEngg']].corr()
```

```python
#2.find the distribution of gender w.r.t to diff classes of 'salary_class' using crosstab?
#a.% of females earning "high income."
#b.% of males earning "high income."

a=pd.crosstab(df.Gender,df.Salary,margins=True)
a
```

```python
a=pd.crosstab([df.DOB],[df.Gender],values=df.Salary,aggfunc=np.average)
a.mean()
```

```python
#who is taken the highest salary and row wise?
print(df['Salary'].max())
df[df.Salary==df.Salary.max()]
```

```python
#.extract the salary class of each candidate followed by their ID, using crosstab()?
a=df.loc[:,'ID':'Salary']
df.pivot_table(index="ID",columns="Salary")
print(a)
```

```python
# find which "collegetier" candidates have good income w.r.to 'salary_class'

df.pivot_table(index="CollegeTier",columns="Salary",margins=True)
```

```python
##how many males or females are attended the amcat?

print(df['Gender'].value_counts()['m'])
print(df['Gender'].value_counts()['f'])

```

```python
#find the frequency distribution of 'specialization' w.r.to different classes of salary_class using crosstab?
pd.crosstab(df.Designation,[ df.Salary,df.JobCity],margins=True)
```
